# Verification of the data subpackage

Reference case for verification: N60 radiation quality, H*(10) operational quantity at 0º irradiation angle, mass energy transfer coefficients for air from PENELOPE 2018, air kerma to dose conversion coefficients from CMI 2025.

Information on N60 radiation quality:

```python
from metpyx.xdata import Qualities

qualities = Qualities()

is_quantity = qualities.is_quality('N60')
series = qualities.get_series('N60')
voltage = qualities.get_voltage('N60')
total_filtration = qualities.get_filtration('N60')
inherent_filtration = qualities.get_filtration('N60', inherent=True)
additional_filtration = qualities.get_filtration('N60', additional=True)
```

```python
print(f'Is standard quality? {is_quantity}')
print(f'Belongs to radiation series: {series}')
print(f'Peak kilovoltage: {voltage} kV')
print(f'Total filtration: {total_filtration} mm Al')
print(f'Inherent filtration: {inherent_filtration} mm')
print(f'Additional filtration: {additional_filtration} mm')
```

```python
assert is_quantity is True
assert series == 'N'
assert voltage == 60
assert total_filtration == {'Al': 4, 'Cu': 0.6}
assert inherent_filtration == {'Al': 4}
assert additional_filtration == {'Cu': 0.6}
print("All verifications passed.")
```

Information on H*(10) operational quantity:

```python
from metpyx.xdata import OperationalQuantities

quantities = OperationalQuantities()

is_quantity = quantities.is_quantity('h_star_10')
is_quantity_angle = quantities.is_quantity_angle('h_star_10', 0)
metadata = quantities.get_quantity('h_star_10')
angles = quantities.get_irradiation_angles('h_star_10')
symbol = quantities.get_symbol('h_star_10')
# mono_h_k = quantities.get_mono_h_k('h_star_10', 0)
```

```python
print(f'Is operational quantity? {is_quantity}')
print(f'Is valid quantity-angle pair? {is_quantity_angle}')
print(f'Metadata: {metadata}')
print(f'Irradiation angles: {angles}')
print(f'Symbol: {symbol}')
```

```python
assert is_quantity is True
assert is_quantity_angle is True
assert metadata == {'symbol': 'H*(10)', 'type': 'ambient', 'depth': 10, 'phantom': None, 'angles': [0]}
assert angles == [0]
assert symbol == 'H*(10)'
print("All verifications passed.")
```
