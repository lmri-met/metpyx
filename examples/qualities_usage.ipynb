# Qualities Usage Example

Script to demonstrate the usage of Qualities class

## Initialization

Import the Qualities class

```python
from data.qualities import Qualities
```

Initialize an XrayQualities object

```python
q = Qualities()
```

## Validation helpers

Check if a series is an x-ray radiation quality series

```python
q.is_series('N')
```

```python
q.is_series('X')
```

Check if a quality is an x-ray radiation quality

```python
q.is_quality('N60')
```

```python
q.is_quality('X60')
```

## Listing radiation qualities and series

List the X-ray radiation quality series

```python
q.series
```

List all the X-ray radiation qualities

```python
q.get_all_qualities()
```

List the qualities in a radiation series

```python
q.get_qualities('H')
```

## Retrieve x-ray tube parameters for a radiation quality

Retrieve the series from a radiation quality

```python
q.get_series('N60')
```

Retrieve the peak kilovoltage from a radiation quality

```python
q.get_voltage('N60')
```

Retrieve the total filtration for a radiation quality

```python
q.get_filtration('N60')
```

Retrieve the inherent filtration for a radiation quality

```python
q.get_filtration('N60', inherent=True)
```

Retrieve the additional filtration for a radiation quality

```python
q.get_filtration('N60', additional=True)
```

## Error handling for invalid quality/series
