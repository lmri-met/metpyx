# OperationalQuantities usage example

Notebook to demonstrate the usage of OperationalQuantities class.

## Initialization

Import the OperationalQuantities class

```python
from data.quantities import OperationalQuantities
```

Initialize an OperationalQuantities object

```python
q = OperationalQuantities()
```

## Validation helpers

Check if a quantity is an operational quantity

```python
q.is_quantity('h_star_10')
```

```python
q.is_quantity('x')
```

Check if a quantity and angle pair is valid

```python
q.is_quantity_angle('h_star_10', 0)
```

```python
q.is_quantity_angle('x', 0)
```

```python
q.is_quantity_angle('h_star_10', 200)
```

## Retrieve information about operational quantities

```python
q.get_all_quantities()
```

```python
q.get_all_quantities(symbol=True)
```

```python
q.get_quantity('h_star_10')
```

```python
q.get_irradiation_angles('h_star_10')
```

```python
q.get_symbol('h_star_10')
```

## Error handling for invalid inputs

```python
try:
    q.get_quantity('x')
except ValueError as e:
    print(f'{e}\n')
```

```python
try:
    q.get_irradiation_angles('x')
except ValueError as e:
    print(f'{e}\n')
```

```python
try:
    q.get_symbol('x')
except ValueError as e:
    print(f'{e}\n')
```
